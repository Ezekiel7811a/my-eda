```python
#Evaluate if linear regression is a good model for the data inputation
data = df.select_dtypes(include=["number"])
independent_variable = data.columns[0]
dependent_variable = data.columns[1]

sns.heatmap(data.corr(), annot=True, cmap='coolwarm')

pearson_df = df[[independent_variable, dependent_variable]].dropna()
pearson_correlation, pearson_p_value = stats.pearsonr(pearson_df[independent_variable], pearson_df[dependent_variable])
print('Pearson Correlation:', pearson_correlation)
```
